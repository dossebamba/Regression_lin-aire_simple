```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Chargement des donn√©es
df = pd.read_csv('../data/homeprices.csv')
df
```

```python
#trac√© du diagramme de dispersion
plt.xlabel('Area(sqt ft)')
plt.ylabel('Price(US$)')
plt.scatter(df['Area'], df['Price'], color ='red', marker = '+')
plt.show()
```

```python
#Cr√©ation d'un objet de r√©gression lin√©aire
reg = LinearRegression()
#ajustement des don√©es
reg.fit(df[['Area']], df['Price'])
```

```python
#Pr√©diction d'une maison avec une supericie de 3300 pieds carr√©s
reg.predict([[3300]])
```

```python
print(reg.coef_)  #valeur du cofficient m
print(reg.intercept_)  #valeur de l'ordonn√©e √† l'origine b
```
