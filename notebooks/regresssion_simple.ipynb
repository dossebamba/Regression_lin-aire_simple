```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Chargement des données
df = pd.read_csv('../data/homeprices.csv')
df
```

```python
#tracé du diagramme de dispersion
plt.xlabel('Area(sqt ft)')
plt.ylabel('Price(US$)')
plt.scatter(df['Area'], df['Price'], color ='red', marker = '+')
plt.show()
```

```python
#Création d'un objet de régression linéaire
reg = LinearRegression()
#ajustement des donées
reg.fit(df[['Area']], df['Price'])
```

```python
#Prédiction d'une maison avec une supericie de 3300 pieds carrés
reg.predict([[3300]])
```

```python
print(reg.coef_)  #valeur du cofficient m
print(reg.intercept_)  #valeur de l'ordonnée à l'origine b
```
